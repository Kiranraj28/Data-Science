# Here's a detailed explanation of the features in the dataset

### EmpNumber
This is the unique identifier for each employee.

### Age
This feature represents the age of the employee.

### Gender
This indicates the gender of the employee.

### EducationBackground
This specifies the employee's educational background (e.g., Life Sciences, Marketing, Medical).

### MaritalStatus
This denotes the employee's marital status (e.g., Married, Divorced, Single).

### EmpDepartment
This indicates the department the employee works in (e.g., Sales, Research & Development, Human Resources).

### EmpJobRole
This specifies the employee's job role within the company (e.g., Sales Executive, Manager, Research Scientist).

### BusinessTravelFrequency
This indicates how frequently the employee travels for business (e.g., Travel_Frequently, Travel_Rarely, Non-Travel).

### DistanceFromHome
This represents the distance (in some unit, likely miles) between the employee's home and workplace.

### EmpEducationLevel
This denotes the employee's education level, with a numerical scale (1 to 5):
- 1 = Below College
- 2 = College
- 3 = Bachelor
- 4 = Master
- 5 = Doctor

### EmpEnvironmentSatisfaction
This measures the employee's satisfaction with their work environment, using a scale of 1 to 4:
- 1 = Low
- 2 = Medium
- 3 = High
- 4 = Very High

### EmpJobInvolvement
This indicates the employee's level of involvement in their job, using a scale from 1 to 4:
- 1 = Low
- 2 = Medium
- 3 = High
- 4 = Very High

### EmpJobSatisfaction
This measures the employee's satisfaction with their job, using a scale from 1 to 4:
- 1 = Low
- 2 = Medium
- 3 = High
- 4 = Very High

### NumCompaniesWorked
This counts the number of companies the employee has worked for.

### OverTime
This is a binary variable indicating whether the employee works overtime (Yes/No).

### EmpLastSalaryHikePercent
This shows the percentage increase in the employee's last salary hike.

### EmpRelationshipSatisfaction
This measures the employee's satisfaction with workplace relationships, using a scale from 1 to 4:
- 1 = Low
- 2 = Medium
- 3 = High
- 4 = Very High

### EmpWorkLifeBalance
This assesses the employee's perception of their work-life balance, using a scale from 1 to 4:
- 1 = Bad
- 2 = Good
- 3 = Better
- 4 = Best

### TotalWorkExperienceInYears
This represents the employee's total work experience in years.

### TrainingTimesLastYear
This indicates the number of times the employee received training in the last year.

### ExperienceYearsAtThisCompany
This shows the number of years the employee has worked at the current company.

### ExperienceYearsInCurrentRole
This indicates the number of years the employee has been in their current job role.

### YearsSinceLastPromotion
This represents the number of years since the employee's last promotion.

### YearsWithCurrManager
This indicates the number of years the employee has been working with their current manager.

### Attrition
This is a binary variable indicating whether the employee has left the company (Yes) or not (No).

### PerformanceRating
This is a rating of the employee's performance:
- 1 = Low
- 2 = Good
- 3 = Excellent
- 4 = Outstanding

### (iii) CATEGORICAL COLUMNS HANDLING

```python
data.info() 
```

- Above statement shows that the 9 categorical features are present in the dataset

```python
cate_fea_name = data.describe(include = ['object', 'category'] ).columns.to_list()[1:] # here we storing without employee id
cate_fea_name   # Extracting the names of categorical features into a list
```

- All the categorical features identified above appear to be of nominal type. Based on this, suitable encoding techniques such as One-Hot Encoding, Frequency Encoding, or Binary Encoding can be applied.

```python
for i in cate_fea_name:  # Extracting the number of unique classes present in each categorical feature
    x = data[i].unique()
    y = len(x)
    print(i,':',y)
```

```python
from sklearn.preprocessing import OneHotEncoder

for col in cate_fea_name:
    if col!='EmpJobRole':
        encoder = OneHotEncoder()
    
    # it gives the output in the form of spare matrix
        encoded = encoder.fit_transform(data[[col]])
    
    #The output of fit_transform() is a sparse array. We convert it to a dense NumPy array using .toarray() and then create a Pandas DataFrame (encoded_df)
        encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out([col]))  
    
    # Here, we join the encoded data with the original data
        data = pd.concat([data.drop(columns=[col]), encoded_df], axis=1)
        
    else:
        mean_performance_by_role = data.groupby('EmpJobRole')['PerformanceRating'].mean()
        data['EmpJobRole'] = data['EmpJobRole'].map(mean_performance_by_role)
```

# (v) HANDLING THE IMBALACE CLASS

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

# 'PerformanceRating' is our target variable and 'X' contains our features
x = data_scaled
y = data['PerformanceRating']

# Initialize SMOTE
smote = SMOTE(random_state=42)  # Set a random state for reproducibility

# Fit and resample the data
X_resampled, y_resampled = smote.fit_resample(x, y)

# 'X_resampled' and 'y_resampled' now contain the oversampled data
# You can check the new class distribution:
y_resampled.value_counts() 
```

**Data Balancing with SMOTE**

After applying **SMOTE** (Synthetic Minority Over-sampling Technique), the dataset has been successfully balanced, effectively addressing the class imbalance issue. This adjustment ensures a more even distribution of the target classes, which is expected to enhance the performance of machine learning models.


# (vi) SPLIT THE DATA

```python
x = X_resampled  # Here, we storing the all input features in x variable
y = y_resampled  # Here, we storing the label features in y variable
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
print(x_train.shape) # here , we can see the No.of input feature used for the training.
print(x_test.shape) # here, we can see the No.of input feature used for the testing.
print(y_train.shape) # here , we can see the No.of output feature used for the training.
print(y_test.shape) # here , we can see the No.of output feature used for the testing.
```

# (vii) MODEL BULIDING
